## Question 6
Write a Python program that modifies a given string by replacing all occurrences of its first character with '$', excluding the first character itself.


Sample String : 'restartâ€˜
Expected Result : 'resta\$t'

```python
# Program to replace all occurrences of the first character with '$'
# (excluding the first character itself)

# Prompt user for input
user_input = input("Enter a string: ")

# Get the first character
first_char = user_input[0]

# Replace occurrences of the first character (ignoring case) in the rest of the string
modified_part = ""
for ch in user_input[1:]:
    if ch.lower() == first_char.lower():
        modified_part += "$"
    else:
        modified_part += ch

# Combine first character with modified part
modified_string = first_char + modified_part

# Display result
print("Modified string:", modified_string)
```

## Question 7

```python
# Program to compute factorial using a while loop

# Prompt user for input
num = int(input("Enter a non-negative integer: "))

# Initialize variables
factorial = 1
i = 1

# Compute factorial using while loop
while i <= num:
    factorial *= i
    i += 1

# Display result
print(f"The factorial of {num} is: {factorial}")
```

## Question 9
Design a Python function to check if a number is prime.


```python
# Simplified function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
num = int(input("Enter a number to check if it is prime: "))
print(f"{num} is prime." if is_prime(num) else f"{num} is not prime.")
```

## Question 11
Design a Python function that takes a tuple of integers as input and returns the largest and smallest elements.

```python
# Function to find largest and smallest elements in a tuple
def min_max(tup):
    if not tup:
        return None, None  # Handle empty tuple
    largest = smallest = tup[0]
    for num in tup[1:]:
        if num > largest:
            largest = num
        if num < smallest:
            smallest = num
    return largest, smallest

# Get input from user
user_input = input("Enter numbers separated by commas: ")
numbers = tuple(int(x.strip()) for x in user_input.split(","))

# Find largest and smallest
largest, smallest = min_max(numbers)

# Display result
print(f"Largest element: {largest}")
print(f"Smallest element: {smallest}")
```

## Question 13
Implement a Python script to identify duplicate elements in a provided array of integers. Return -1 if there are no duplicates.

```python
# Program to find duplicates in a list of integers

# Get input from user (comma-separated integers)
user_input = input("Enter numbers separated by commas: ")
num_list = [int(x.strip()) for x in user_input.split(",")]

# Dictionary to count occurrences
count_dict = {}
for num in num_list:
    count_dict[num] = count_dict.get(num, 0) + 1

# Find duplicates
duplicates = [num for num, count in count_dict.items() if count > 1]

# Display result
if duplicates:
    print("Duplicate elements:", duplicates)
else:
    print(-1)
```

```python
# Program to count even numbers from 1 to n using continue

# Get input from user
n = int(input("Enter a positive integer: "))

# Initialize counter
even_count = 0

# Loop through numbers from 1 to n
for i in range(1, n + 1):
    if i % 2 != 0:
        continue  # Skip odd numbers
    even_count += 1

# Display result
print(f"Count of even numbers from 1 to {n}: {even_count}")
```

## Question 17
Develop a Python function called 'count_vowels' that takes a string as input and returns the count of vowels (both lowercase and uppercase) present in the string, ignoring spaces and non-alphabetic characters.

```python
# Function to count vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char.isalpha() and char in vowels:
            count += 1
    return count

# Example usage
user_input = input("Enter a string: ")
vowel_count = count_vowels(user_input)
print(f"Number of vowels in the string: {vowel_count}")
```

## Question 20
Implement a Python script to generate a random password of a specified length.

```python
import random
import string

# Function to generate random password
def generate_password(length):
    if length < 4:
        return "Password length should be at least 4 for complexity."

    # Character pools
    letters = string.ascii_letters   # a-z, A-Z
    digits = string.digits           # 0-9
    symbols = string.punctuation     # Special characters

    # Ensure password has at least one character from each category
    password = [
        random.choice(letters),
        random.choice(digits),
        random.choice(symbols)
    ]

    # Fill the remaining length with random choices from all characters
    all_chars = letters + digits + symbols
    password += random.choices(all_chars, k=length - 3)

    # Shuffle the password list to randomize order
    random.shuffle(password)

    # Convert list to string
    return ''.join(password)

# Get password length from user
length = int(input("Enter desired password length: "))
print("Generated password:", generate_password(length))
```

(21) Define a base class called 'Vehicle' with attributes make and model. Create a derived class Car that inherits from Vehicle and has an additional attribute 'num_doors'. Demonstrate an example of creating an instance of the 'Car' class and accessing its attributes.

```python
# Base class
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

# Derived class
class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)  # Call base class constructor
        self.num_doors = num_doors

# Example: creating an instance of Car
my_car = Car("Toyota", "Corolla", 4)

# Accessing attributes
print("Make:", my_car.make)
print("Model:", my_car.model)
print("Number of doors:", my_car.num_doors)
```

(22) Create a class called 'Shape' with a method calculate_area(). Implement two derived classes 'Circle' and 'Rectangle' that inherit from Shape and override the calculate_area() method. Demonstrate polymorphism by creating instances of both derived classes and calling the calculate_area() method.

```python
import math

# Base class
class Shape:
    def calculate_area(self):
        raise NotImplementedError("This method should be overridden by subclasses")

# Derived class: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

# Derived class: Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

# Demonstrate polymorphism
shapes = [
    Circle(7),         # Circle with radius 5
    Rectangle(8, 9)    # Rectangle with length 4 and width 6
]

for shape in shapes:
    print(f"Area of {shape.__class__.__name__}: {shape.calculate_area()}")
```
